````python
# to download tiled images as well as a whole image, we should set '--tile' as true
# '--tile-width' and '--tile-height' decide tiled image's size
# '--mp=true' is for multiprocessing which speeds up the download
# if you only want to download original images, delete tile related parameters from the command.

!gdl 2021-08-01 2021-08-15 "24.721263153548602, -84.10379404218443" "30.594775921079584, -79.46756349558221" --tile=true --tile-width=512 --tile-height=512 --mp=true
````
